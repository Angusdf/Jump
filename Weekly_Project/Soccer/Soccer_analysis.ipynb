```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#1. Load the csv file and show top 5 records from it
df= pd.DataFrame(pd.read_csv('/content/players_20.csv'))
df.head()
```

```python
#2. How you would be able to see each column's name.
for col in df.columns:
  print(col)
#Row x Column 3. Need to show number of rows and columns of this dataset.
df.shape
```

```python
#4. Show number of players and their countries
country_count = pd.DataFrame(df.groupby('nationality')['short_name'].count()).rename(columns={'short_name': 'Count'}).sort_values('Count')
country_count
```

```python
#5. If you find many records in point 4 then show only top 10 countries and their
#number of players.
country_count.tail(10)
```

```python
#6. Now you have to create a bar plot of top 5 countries and their number of
#players, try to fill green color in bars.
country_count.tail(5).plot.bar(color={"Count":"green"}, figsize=(20,10), title="Nation Count")
```

```python
#7. Show top 5 players short name and wages
df[['short_name','wage_eur']].head(5)
```

```python
#8. Show top 5 players short name and wages that are getting highest salaries.
high_player_wage = df[['short_name','wage_eur']].sort_values(by='wage_eur',ascending=False).head(5)
high_player_wage
```

```python
#9. Create a bar plot of point number 8
high_player_wage.plot.bar(x='short_name',figsize=(20,10), title='Highest Player Wage in Euros')
```

```python
# Is there a correlation between: age, height, weight, pay, potential, value, rating  
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(df[['age','height_cm','potential','wage_eur', 'value_eur','overall']].corr())
```

```python
sns.set(rc={'figure.figsize':(20,20)})
pd.plotting.scatter_matrix(df[['age','height_cm','potential','wage_eur', 'value_eur','overall']])
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(df[['pace','shooting','passing','dribbling', 'defending','physic', 'potential','wage_eur']].corr())
```

```python
#average pay by postion
df[['team_position','wage_eur']].groupby(df['team_position'])\
                                .mean().sort_values('wage_eur',ascending= False)\
                                .plot.bar(title='Average Pay per Position')
```

```python
df[['club','value_eur']].groupby(df['club'])\
                        .sum().sort_values('value_eur',ascending= False)\
                        .head(30).plot.bar(title="Value of Club per Club")
```
