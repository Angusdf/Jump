```python
import pandas as pd
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Load the CSV into a dataframe printing the top 5 to view data with in
df= pd.DataFrame(pd.read_csv('/content/drive//MyDrive/code/Data_Set/charcters_stats.csv'))
df.head()
```

```python
#Show the  number of Rows and Columns
df.shape
```

```python
#Show the count of each Alignment 
df.Alignment.value_counts()
```

```python
#Show the top 5 good superheros
df.loc[df.Alignment == 'good'].head()
```

```python
#Show top 5 speedy superheros 
df.sort_values(by=['Speed'], ascending = False)\
  .loc[df.Alignment == 'good'].head()
```

```python
#Show top 5 power superheros 
df.sort_values(by=['Power'], ascending = False)\
  .loc[df.Alignment == 'good'].head()
```

```python
#How many good superheroes have max power
df.loc[(df.Power== 100) & (df.Alignment == 'good')]['Power'].count()
```

```python
#shape of the dataframe of heroes 
good_max_power = df.loc[(df.Power== 100) & (df.Alignment == 'good')]
good_max_power.shape
```

```python
#Show good superheroes have max power
good_max_power=good_max_power.sort_values('Total',ascending = False)
good_max_power
```

```python
# Get the top five of the good supers with max power
good_max_power.head()
```

```python
#display the stats of the top five heros
good_max_power.head().plot(kind='bar', x='Name', figsize=(10,5)
                    ,title= "Top Five Good Hero Stats", rot = 75)
```

```python
#Display the 'total' good heroes in green top five 
good_max_power.head()[['Name','Total']].plot(kind='bar', x='Name', figsize=(10,5)
                    ,title= "Top Five Good Hero Stats", color = 'green')
```

```python
#Gather all villians
villians_df = df.loc[df.Alignment == 'bad']
villians_df
```

```python
#show top five from abouve 
villians_df.head()
```

```python
#show top five fastest villains 
villians_df[['Name','Speed']].sort_values('Speed', ascending = False).head()
```

```python
#show top five intelligent villains 
villians_df[['Name','Intelligence']].sort_values('Intelligence', ascending = False).head()
```

```python
#Show most dangerous villians by total 
villians_df[['Name','Total']].sort_values('Total', ascending = False).head()
```

```python
hero_df = df.loc[df.Alignment == 'good']
#set the figure size to desired size
sns.set(rc={'figure.figsize':(10,5)})
#histogram of speed from heros
sns.histplot(data=hero_df, x='Speed', shrink=.9).set(title='Distribution of Speed')
```

```python
#histogram of combat from villians
sns.histplot(data=villians_df, x='Combat', shrink=.9).set(title='Distribution of Combat')
```

```python
#Pie graph showing alignments
df.Alignment.value_counts().plot.pie(title = 'Graph of Alignment',
                                      figsize = (20,10),
                                     autopct='%1.1f%%', fontsize=30, explode=(.05,.05,.05))
```

```python
#Is there a corraltion over all data?
sns.heatmap(df.corr(method='pearson')).set(title = "Heatmap of the Data Set")
```

```python
heroes_df = df.loc[df.Alignment == 'good']
#Is there a corraltion heros data?
sns.heatmap(heroes_df.corr(method='pearson')).set(title = 'Heatmap of Heroes')
```

```python
#Is there a corraltion villians data?
sns.heatmap(villians_df.corr(method='pearson')).set(title = 'Heatmap of Villians')
```

```python
#Is there a corraltion neutral data?
sns.heatmap(df.loc[df.Alignment == 'neutral'].corr(method='pearson')).set(title = 'Heatmap of Netural')
```

```python
#What are the stats of the top five villians and heros, sorted by total?
villians_df.sort_values('Total', ascending = False).head()[['Name',	
                                                          'Intelligence',	
                                                          'Strength', 
                                                          'Speed',	
                                                          'Durability',	
                                                          'Power', 
                                                          'Combat']]\
                                                          .plot(kind='bar', x='Name', figsize=(10,5)
                    ,title= "Top Five Villians Stats")
villians_df.sort_values('Total', ascending = False).head()
```

```python
#What are the stats of the top five villians and heros, sorted by total?
df.loc[df.Alignment == 'neutral'].sort_values('Total', ascending = False).head()[['Name',	
                                                          'Intelligence',	
                                                          'Strength', 
                                                          'Speed',	
                                                          'Durability',	
                                                          'Power', 
                                                          'Combat']]\
                                                          .plot(kind='bar', x='Name', figsize=(10,5)
                    ,title= "Top Five Neutral Stats")
df.loc[df.Alignment == 'neutral'].sort_values('Total', ascending = False).head()
```
