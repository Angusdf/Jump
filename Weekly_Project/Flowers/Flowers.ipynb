# Applying Linear Regression

```python
y= iris_df.sepal_length
x= iris_df.sepal_width
```

```python
#Traing selecting 'Random' data to be used, test_ proportion of the dataset to include in the train split
x_train, x_test, y_train, y_test = model_selectionms.train_test_split(x,y,test_size=.3)
```

```python
#converting to Dataframes for easier use
x_train=pd.DataFrame(x_train)
y_train=pd.DataFrame(y_train)
x_test=pd.DataFrame(x_test)
y_test=pd.DataFrame(y_test)
x_train.head()
```

```python

print('x_train\n',pd.DataFrame(x_train).head().to_markdown())
print('x_test\n',pd.DataFrame(x_test).head().to_markdown())
print('y_train\n',pd.DataFrame(y_train).head().to_markdown())
print('y_test\n',pd.DataFrame(y_test).head().to_markdown())
```

```python
#Traning the ML
lr = LinearRegression()
lr.fit(x_train, y_train)
print('Score: ',lr.score(x_train, y_train)) #Return the coefficient of determination of the prediction. Closer to one is good 
print('Coef: ', lr.coef_) #Return the coefficient of determination of the prediction. (Slope)
print('Intercept: ',lr.intercept_)
```

```python
#Predicting the y
y_pred = lr.predict(x_test)
y_pred[0:5]
```

```python
#Testing the y_predicted to y_test Closer to zero is best
model_1_mse = mean_squared_error(y_test.sepal_length,y_pred.flatten().tolist())
model_1_mse
```

```python
#Ploting the predicted vs the test on the relationship graph of Sepal Width vs Sepal Length
sns.scatterplot(data=iris_df, x='sepal_width', y='sepal_length',  hue ='species')\
  .set(title = 'Sepal Width vs Sepal Length')
sns.lineplot(x=x_test.sepal_width, y=y_pred.flatten().tolist())
```

Shows the relationship between Petal Lenght and Sepal Length over the complete data set. The figure also shows the Linear Regression line.

```python
from sklearn import svm
import sklearn.metrics as metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import PrecisionRecallDisplay
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
import numpy as np
from sklearn.preprocessing import normalize
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
sns.heatmap(iris_df.corr(), annot = True) .set(title = 'Correlation of the Iris Data Set')
```

Showing the corelation of the data set. Closer to one/ lihgter color the higher the correlation

```python
#The fit time scales at least quadratically with the number of samples 
#and may be impractical beyond tens of thousands of samples
y_extra = iris_df.species
x_extra = iris_df[['sepal_width', 'petal_length', 'petal_width','sepal_length']]
model_SVC = svm.SVC()
x_train_extra, x_test_extra, y_train_extra, y_test_extra = model_selectionms.train_test_split(x_extra,y_extra,test_size=.3)
model_SVC.fit(x_train_extra, y_train_extra)
y_pred_SVC = model_SVC.predict(x_test_extra)
print('Accuracy: ',metrics.accuracy_score(y_test_extra.tolist(), y_pred_SVC))
print(metrics.classification_report(y_test_extra.tolist(), y_pred_SVC))
#precison - ratio tp / (tp + fp) where tp is the number of true positives 
#and fp the number of  positives.does depend on the classifier threshold
#recall - ratio tp / (tp + fn) where tp is the number of true positives 
#and fn the number of false positives. does not depend on the classifier threshold
#f1-score - defined as the harmonic mean of precision and recall. 2 (P*R)/(P+R)
#support - is the number of occurrences of each class in y_true
#macro avg - Macro averaging reduces your multiclass predictions down to 
#   multiple sets of binary predictions, calculates the corresponding metric 
#   for each of the binary cases, and then averages the results together. 

#The diagonal elements represent the number of points for which the predicted
# label is equal to the true label, while off-diagonal elements are those that 
# are mislabeled by the classifier. The higher the diagonal values of the 
# confusion matrix the better, indicating many correct predictions.
disp = metrics.ConfusionMatrixDisplay(metrics.confusion_matrix(y_test_extra,y_pred_SVC),
                                      display_labels=['setosa', 'versicolor', 'virginica'],
                                      )
disp.plot()
disp.ax_.set_title('Confusion Matrix of a C-Support Vector Classification')
plt.show()
acc = {}
acc ['SVC'] = metrics.accuracy_score(y_test_extra.tolist(), y_pred_SVC)
```

The Confusion Matrix of a Confusion Matrix of C-Support Vector Classification. 

```python
model_LogReg = LogisticRegression()

model_LogReg.fit(x_train_extra, y_train_extra )
y_pred_LogReg = model_LogReg.predict(x_test_extra)
print('Accuracy: ',metrics.accuracy_score(y_test_extra.tolist(), y_pred_LogReg))
print(metrics.classification_report(y_test_extra.tolist(), y_pred_LogReg))
 
disp = metrics.ConfusionMatrixDisplay(metrics.confusion_matrix(y_test_extra,y_pred_LogReg),
                                      display_labels=['setosa', 'versicolor', 'virginica'],
                                      )
disp.plot()
disp.ax_.set_title('Confusion Matrix of Linear Regression Classification')
plt.show()
acc ['LogisticRegression'] = metrics.accuracy_score(y_test_extra.tolist(), y_pred_LogReg)
```

The Confusion Matrix of a Confusion Matrix of Linear Regression Classification. 

```python
model = DecisionTreeClassifier()
model.fit(x_train_extra, y_train_extra )
y_pred_tree = model.predict(x_test_extra)

print('Accuracy: ',metrics.accuracy_score(y_test_extra.tolist(), y_pred_tree))
print(metrics.classification_report(y_test_extra.tolist(), y_pred_tree))


disp = metrics.ConfusionMatrixDisplay(metrics.confusion_matrix(y_test_extra,y_pred_tree),
                                      display_labels=['setosa', 'versicolor', 'virginica'],
                                      )
disp.plot()
disp.ax_.set_title('Confusion Matrix of Decision Tree Classification')
plt.show()
acc ['DecisionTree'] = metrics.accuracy_score(y_test_extra.tolist(), y_pred_tree)
```

The Confusion Matrix of Decision Tree Classification.

```python
model = KNeighborsClassifier(n_neighbors=6)
model.fit(x_train_extra, y_train_extra )
y_pred_KN = model.predict(x_test_extra)

print('Accuracy: ',metrics.accuracy_score(y_test_extra.tolist(), y_pred_KN))
print(metrics.classification_report(y_test_extra.tolist(), y_pred_KN))


disp = metrics.ConfusionMatrixDisplay(metrics.confusion_matrix(y_test_extra,y_pred_KN),
                                      display_labels=['setosa', 'versicolor', 'virginica'],
                                      )
disp.plot()
disp.ax_.set_title( 'Confusion Matrix of KNeighbor Classification.')
plt.show()

acc [' KNeighbors'] = metrics.accuracy_score(y_test_extra.tolist(), y_pred_KN)
```

The Confusion Matrix of KNeighbor Classification

```python
plt.barh(list(acc), list(acc.values()))
for index, value in enumerate(list(acc.values())):
  plt.text(value, index, str(value))
plt.title('Accuracy Comparison of ML Classification')
plt.show()
```
